```python
import numpy as np # generic math functions
import matplotlib.pyplot as plt
import time
import scipy
```

```python
N_min, N_max = (5, 9)

delta_E_matrix = np.zeros((N_max-N_min,20))
expval_matrix = np.zeros((N_max-N_min,20))

for i, N in enumerate(range(N_min, N_max)):
    delta_E_matrix[i, :] = np.load(r'/media/maciejko/zhan/top/data/delta_E_N_para'+str(N)+'.npy')
    expval_matrix[i, :] = np.load(r'/media/maciejko/zhan/top/data/expval_N_para'+str(N)+'.npy')
    
delta_E_matrix = abs(delta_E_matrix)
expval_matrix = abs(expval_matrix)
```

```python
for i, N in enumerate(range(N_min, N_max)):
    plt.plot(np.arange(0,2,step=0.1),delta_E_matrix[i,:])
    plt.scatter(np.arange(0,2,step=0.1),delta_E_matrix[i,:],label='N='+str(N))
plt.xlabel('J/h')
plt.ylabel('(E_1 - E_0)')
plt.title('Lowest state transition as a function of J/h')
plt.legend()
plt.savefig('/media/maciejko/zhan/top/plots/Band Gap.png')
plt.savefig('/media/maciejko/zhan/top/plots/Band Gap.pdf')
```

```python
for i, N in enumerate(range(N_min, N_max)):
    plt.plot(np.arange(0,2,step=0.1),expval_matrix[i,:])
    plt.scatter(np.arange(0,2,step=0.1),expval_matrix[i,:],label='N='+str(N))
plt.xlabel('J/h')
plt.ylabel('Expval of M^2')
plt.ylim(bottom=0)
plt.title('Expval of M^2')
plt.legend()
plt.savefig('/media/maciejko/zhan/top/plots/Expval.png')
plt.savefig('/media/maciejko/zhan/top/plots/Expval.pdf')
```
